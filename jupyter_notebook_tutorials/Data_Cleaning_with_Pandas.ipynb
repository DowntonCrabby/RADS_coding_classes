**use the  [.str.split()](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html) to split the "City, State" column into "City" and "State" columns**

* view the data in the dataframe to see what the separator is
* split the string into two separate columns
    * df['column name'].str.split(separator, Limit number of splits in output, list output in separate columns)  
* drop the old column  with the unsplit data
* add the split columns to the datframe

```python
# view the "City, State" column, to see the separator
df["City, State"]
```

```python
# Using , as the separator we will create a new new_string_df containing those two columns
split_string_df = df["City, State"].str.split(",", n = 1, expand = True)

# view the results
split_string_df
```

```python

df["City, State"]
```

```python
# Here we will use the df.str.split() function to split the "City, State" column into "City" and "State" 
# Using , as the separator we will create a new new_string_df containing those two columns
new_string_df = df["City, State"].str.split(",", n = 1, expand = True)

# view the results
new_string_df
```

```python
# We will also drop the original "City, State" from the original df
# Rows are axis 0 so we want to be sure to drop from axis 1 for columns
# We will name the Dataframe with the dropped column df_spil
df_split = df.drop("City, State",axis=1)

# view the results, is "City, State" still a column?
df_split.columns
```

```python
# We will then add "City" and "State" as columns into our df_split
df_split["City"]= new_string_df[0]
df_split["State"]= new_string_df[1]

# view the results, are the new columns added?
df_split.columns
```

```python
# Using df['column'] allows us to pull a particular column as a pandas Series datatype
# Using df[['column']] allows us to pull it as a Dataframe
# multiple columns can be selected this way separated by a comma df['col1', 'col2']
df_split['City']
```

```python
# We will drop the "Category (OLD)" column and call the new Dataframe df_drop
df_drop = df_split.drop('Category (OLD)',axis=1)
```

```python
# Next we will filter the data to just the USA and call the new Dataframe df_filter
df_filter = df_drop.loc[df['Country'] == "United States"]
```

```python
# Here we can view df_filter
df_filter
```

<div style="padding: 5px; padding-left: 10px; background: #e6f2ff">
<h2> Table Manipulation </h2> 

* pivot

```python
# You may have noticed that there are many missing values in the columns that contain each month, ie "1/1/2014"
# We will pivot these values from columns to rows, creating two new columns, one for the column 12 column headers
# And one for the values.
df_pivot = pd.melt(df_filter,id_vars=['Category', 'Country', 'Customer ID', 'Customer Name', 'Discount',
'Market', 'Order ID', 'Order Priority', 'Postal Code', 'Product ID',
'Product Name', 'Profit', 'Quantity', 'Region', 'Row ID', 'Segment',
'Ship Mode', 'Shipping Cost', 'Sub-Category',
'City', 'State'], value_vars=['10/1/2014', '7/1/2014', '11/1/2014', '9/1/2014',
'1/1/2014', '12/1/2014', '8/1/2014', '5/1/2014', '3/1/2014', '4/1/2014',
'2/1/2014', '6/1/2014'],var_name='Order Date', value_name='Sales')
```

```python
# We can now see the new Dataframe, df_pivot
df_pivot
```

```python
# Using the .dropna() function we can remove rows with missing values
# inplace=True does this to the existing Dataframe without needing to create a new one
df_pivot.dropna(subset=['Sales'],inplace=True)
```

```python
df_pivot
```

```python
# If you noticed above, Postal Code is being read as a float type, but we can convert it to in using .astype()
df_pivot['Postal Code'] = df_pivot['Postal Code'].astype(int)
```

```python
# We can rename the Order Priority column to Urgency using the .rename() function
df_pivot.rename(columns={"Order Priority": "Urgency"}, inplace=True)
```

```python
# We can create a new field showing the Cost of Goods Sold or COGS by subracting the "Profit" column from "Sales"
# Pandas does this efficiently using vectorized operations, so we don't need to write a loop
df_pivot['COGS'] = df_pivot['Sales'] - df_pivot['Profit']
```

```python
# We can un-pivot rows into columns using the .pivot() function as well
df_un_pivot = df_pivot[['State', 'Category']].pivot( columns='Category')
```

```python
df_un_pivot
```

```python
# the .pivot_table() function allows us to construct quick aggregate tables using categorical variables
pd.pivot_table(df_pivot, values =['Order ID'], index=['Urgency'], columns=['Ship Mode'], aggfunc='count')
```

```python
# Pandas has a set of charting tools built into the library.
# By aggregating data using df.grouby().agg() we can quickly construct visualizations, for example here using df.plot.bar()
# Alternatively we could use df.plot(type='bar')
df_pivot[['Region', 'COGS']].groupby('Region').agg(['sum']).plot.bar()
plt.xlabel('Region')
plt.ylabel('COGS')
```

```python
# We can extend the aggregation by sorting the data using df.sort_values(by=[('column', 'aggregation type')])
df_agg = df_pivot[['State', 'Sales']].groupby('State').agg(['sum'])
df_agg.sort_values(by=[('Sales', 'sum')], ascending = False, inplace=True)

# figsize can be used to increase the size of the visualization
df_agg.plot.barh(figsize=(10,10))
```

### Student Activity

1) What State has the highest Profit in the US?

```python
# aggregate the correct data & sort it
profit_by_state = df_pivot[['State', 'Profit']].groupby('State').agg(['sum'])
profit_by_state.sort_values(by=[('Profit', 'sum')], ascending = False, inplace=True)

# print state with highest profit
print("Most profitable state: " + profit_by_state.index[0])
```

2) Which Sub-Category has the most Quantity of products ordered?

```python
# same method to aggregate and sort the data
subcat_quant = df_pivot[['Sub-Category', 'Quantity']].groupby('Sub-Category').agg(['sum'])
subcat_quant.sort_values(by=[('Quantity', 'sum')], ascending = False, inplace=True)

# again print results 
print("Sub-Category with the largest quantity: " + subcat_quant.index[0])
```

3) Which Customer had the 3rd highest total Sales?

```python
# same process as above
sales_by_cust = df_pivot[['Customer Name', 'Sales']].groupby('Customer Name').agg(['sum'])
sales_by_cust.sort_values(by=[('Sales', 'sum')], ascending = False, inplace=True)

# get the 3rd row
print("Customer with 3rd highest total sales: " + sales_by_cust.index[2])
```
